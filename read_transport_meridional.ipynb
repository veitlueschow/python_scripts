# Read meridional transport and make it look nice

```python
from netCDF4 import Dataset
import os
import numpy as np
import matplotlib.pyplot as plt
from imp import reload
import tools
import numpy.ma as ma
```

```python
path = "/work/mh0256/m300522/data_storm/amoc/calc/30N35S"
os.chdir(path)
```

```python
filename = 'transport_meridional_full3.nc'
fh= Dataset(filename,mode='r')
full_ = fh.variables["transport_mer_full"]
full  = full_[:,:].copy()

y_ = fh.variables["lat"]
y  = y_[:,:].copy()
z_ = fh.variables["depth"]
z  = z_[:,:].copy()

filename = 'transport_meridional_dwbc3.nc'
fh= Dataset(filename,mode='r')
dwbc_ = fh.variables["transport_mer_dwbc"]
dwbc  = dwbc_[:,:].copy()
```

```python
z.shape
```

```python
#fig, axes = plt.subplots(nrows=1, ncols=2)

plt.figure(figsize=(10,5))
high=2.1e7
v = np.linspace(-high,high,100)
plt.contourf(y,-z,dwbc,v,extend="both")
plt.xlabel("Latitude in deg",fontsize=17)
plt.ylabel("Depth in m",fontsize=17)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.set_cmap('RdBu_r')

plt.savefig("/work/mh0256/m300522/data_storm/amoc/figs/dwbc_transport.png",bbox_inches = 'tight')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.contourf(y,-z,full,v,extend="both")
plt.xlabel("Latitude in deg",fontsize=17)
plt.ylabel("Depth in m",fontsize=17)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
cb =plt.colorbar(format='%.0e',ticks=[-2e7,-1e7, 0, 1e7,2e7])
cb.ax.tick_params(labelsize=15) 
cb.set_label("m$^3$/s", fontsize=17)
#plt.savefig("/work/mh0256/m300522/data_storm/amoc/figs/full_transport.png",bbox_inches = 'tight')
plt.show()
```
