# Plot geostrophic ratio in 1941.5 m

```python
from netCDF4 import Dataset
import os
import numpy as np
import matplotlib.pyplot as plt
from imp import reload
import tools
import numpy.ma as ma
```

```python
#path="/work/mh0256/m300522/data_storm/geostrophy/geostrophy_mean"
path="/work/mh0256/m300522/data_storm/geostrophy/2000s"
os.chdir(path)
```

```python
filename = 'plot.nc'
fh= Dataset(filename,mode='r')
raw_ = fh.variables["jochems_raw"]
raw  = raw_[0,0,:,:].copy()

#filename = 'plot_emfd.nc'
#fh= Dataset(filename,mode='r')
#emfd_ = fh.variables["jochems_emfd"]
#emfd  = emfd_[0,0,:,:].copy()

#filename = 'plot_emfd_mma.nc'
#fh= Dataset(filename,mode='r')
#mma_ = fh.variables["jochems_emfd_mma"]
#mma  = mma_[0,0,:,:].copy()

y_ = fh.variables["lat"]
y  = y_[:,:].copy()
x_ = fh.variables["lon"]
x  = x_[:,:].copy()
```

```python
v=np.linspace(0,1,10)
fig=plt.figure(figsize=(6,10))
a=fig.add_subplot(111,axisbg='lightgray')
plt.contourf(x[190:1000,100:600],y[190:1000,100:600],raw[190:1000,100:600],v,extend="both")
plt.xlabel("Longitude in deg",fontsize=15)
plt.ylabel("Latitude in deg",fontsize=15)
plt.set_cmap("Blues")
plt.colorbar(ticks=[0,0.2,0.4,0.6,0.8,1,])
plt.savefig("geostrophic_error_1941.5.png",bbox_inches = 'tight')
#cmap.set_bad('black',1.)
plt.show()
```
